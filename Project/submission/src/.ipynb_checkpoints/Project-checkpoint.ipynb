```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def contrastStretching(img):
    minPixel = img.min()
    maxPixel = img.max()
    
    P1 = np.subtract(img, minPixel)
    P2 = np.multiply(P1, 255/(maxPixel - minPixel))

    return P2
```

## Otsu Thresholding

```python
img1 = cv2.imread('sample01.png',0)
ret, thresh1 = cv2.threshold(img1, 120, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)      
print(ret)

plt.figure(figsize=(10,10))
plt.imshow(thresh1, cmap = 'gray')
plt.title('Otsu')

# cv2.imwrite('sample01_otsu.png', thresh1)
```

```python
img2 = cv2.imread('sample02.png',0)
ret, thresh2 = cv2.threshold(img2, 120, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)      
print(ret)

plt.figure(figsize=(7,7))
plt.imshow(thresh2, cmap = 'gray')
plt.title('Otsu')

# cv2.imwrite('sample02_otsu.png', thresh2)
```

## Canny Edge Detection

```python
edges1 = cv2.Canny(img1,100,200)

plt.figure(figsize=(10,10))
plt.imshow(edges1, cmap = 'gray')
plt.title('Canny Edge')

# cv2.imwrite('sample01_canny.png', edges1)
```

```python
edges2 = cv2.Canny(img2,100,200)

plt.figure(figsize=(7,7))
plt.imshow(edges2, cmap = 'gray')
plt.title('Canny Edge')

# cv2.imwrite('sample02_canny.png', edges2)
```

## Contrast Stretching

```python
cs1 = contrastStretching(img1)

plt.figure(figsize=(10,10))
plt.imshow(cs1, cmap='gray')
# cv2.imwrite('sample01_cs.png', cs1)
```

```python
cs2 = contrastStretching(img2)

plt.figure(figsize=(7,7))
plt.imshow(cs2, cmap='gray')
# cv2.imwrite('sample02_cs.png', cs2)
```

## Medium Blur to Improve Image 1

```python
blured1_61 = cv2.medianBlur(img1, 1)
blured2_61 = cv2.medianBlur(img1, 61)

divided_61 = np.ma.divide(blured1_61, blured2_61).data
normed_61 = np.uint8(255 * divided_61 / divided_61.max())

th, threshed_61 = cv2.threshold(normed_61, 100, 255, cv2.THRESH_OTSU)
plt.figure(figsize=(10,10))
plt.imshow(threshed_61, cmap='gray')

# cv2.imwrite('sample01_MB_thres61.png', threshed_61)
```
